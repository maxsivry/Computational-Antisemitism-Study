```python
import pandas as pd

# Assuming your CSV files are named file1.csv, file2.csv, and file3.csv
file_paths = ['M.csv', 'T.csv', 'J.csv', 'S.csv']

# Create an empty DataFrame to store the merged data
merged_df = pd.DataFrame()

# Iterate through each file and append its data to the merged DataFrame
for file_path in file_paths:
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)
    
    # Drop rows with blank "label" values
    df = df.dropna(subset=['label'])
    
    # Append the current DataFrame to the merged DataFrame
    merged_df = merged_df.append(df, ignore_index=True)

print(merged_df)
print((len(merged_df[merged_df['label'] == 1])/ merged_df.shape[0]) * 100)

merged_df.to_csv('training.csv', index=False)
```
